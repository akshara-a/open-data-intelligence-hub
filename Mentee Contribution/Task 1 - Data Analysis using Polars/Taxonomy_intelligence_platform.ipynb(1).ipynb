```python
# ==============================================================================
# ENTERPRISE GEOSPATIAL TAXONOMY INTELLIGENCE & ANALYTICS PIPELINE SYSTEM
# PRODUCTION ARCHITECTURE: UNIFIED OBJECT-ORIENTED INDUSTRIAL SYSTEM BLUEPRINT
# ==============================================================================

# STAGE 01: DEPENDENCY INJECTION & SYSTEM MODULE INSTALLATION
print("[SYSTEM INITIALIZATION] Setting up enterprise execution ecosystem...")
import sys
import subprocess

# Auto-verify and install core packages dynamically to prevent initialization failures
required_packages = ["polars", "pyarrow", "scikit-learn", "geopy", "gradio"]
for package in required_packages:
    try:
        __import__(package)
    except ImportError:
        print(f"[ENVIRONMENT INFO] Embedding package framework: {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package, "--quiet"])

# Core Imports Execution
import os
import polars as pl
import numpy as np
import gradio as gr
from geopy.geocoders import Nominatim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# PIPELINE PARAMETERS CONFIGURATION
#  CSV file ka correct path :
TARGET_DATASET_CSV = "/content/drive/MyDrive/taxonomy_file_data.csv"
SAMPLING_FRACTION = 0.15  # Standard 10% balanced sample slice execution

print("\n[STAGE 01/05] Launching High-Performance Out-of-Core Polars Core Data Pipeline...")
if not os.path.exists(TARGET_DATASET_CSV):
    raise FileNotFoundError(f"Configuration Critical Error: Source verification failed for path: '{TARGET_DATASET_CSV}'")

# 1. Zero-RAM Lazy Evaluation Matrix Setup
lazy_df = pl.scan_csv(TARGET_DATASET_CSV, separator="\t", infer_schema_length=10000, ignore_errors=True)

# 2. Schema Scrubbing and Category Optimization Filter
processed_lazy = (
    lazy_df.select([
        pl.col("decimalLatitude"),
        pl.col("decimalLongitude"),
        pl.col("kingdom"),
        pl.col("phylum"),
        pl.col("class"),
        pl.col("basisOfRecord"),
        pl.col("year"),
        pl.col("month")
    ])
    .drop_nulls()
    .filter(pl.col("kingdom").is_in(["Animalia", "Plantae", "Fungi"]))
)

# 3. Dynamic Feature Engineering: Absolute Latitudinal Gradient Integration
processed_lazy = processed_lazy.with_columns(
    pl.col("decimalLatitude").abs().alias("absolute_latitude")
)

# 4. Compiling Polars Streaming Core Pipe
df_platform = processed_lazy.collect(streaming=True).sample(fraction=SAMPLING_FRACTION, seed=42)

# 5. Extract Analytics KPI Indicators via Polars Engine
total_records = df_platform.shape[0]
most_common_kingdom = df_platform.group_by("kingdom").agg(pl.len()).sort("len", descending=True)["kingdom"][0]
top_phylum = df_platform.group_by("phylum").agg(pl.len()).sort("len", descending=True)["phylum"][0]

print("🌍 AUTOMATED ENTERPRISE GEOSPATIAL ANALYTICS GENERATOR REPORT")
print(f"1. DATA COMPLIANCE   : Successfully parsed and compiled {total_records:,} metadata rows.")
print(f"2. TAXONOMICAL BIAS  : Primary core ecological distribution cluster: '{most_common_kingdom}'.")
print(f"3. GRANULAR DENSITY  : Dominant dispersion node identified within Phylum: '{top_phylum}'.")
print(f"4. PIPELINE STATUS   : Engineered absolute space-time matrix modules successfully.")
print("==========================================================================\n")


print("[STAGE 02/05] Executing Categorical Label Encoding & Dataset Stratification...")
# 1. Feature Ingestion Arrays Transformation
X_data = df_platform.select(["decimalLatitude", "decimalLongitude", "month", "absolute_latitude"]).to_numpy()
y_strings = df_platform.select("kingdom").to_numpy().ravel()

# 2. Transforming target labels into machine-readable format
label_encoder = LabelEncoder()
y_data = label_encoder.fit_transform(y_strings)
classes_list = label_encoder.classes_

# 3. Stratified Train-Test Boundary Configuration (80% Train, 20% Evaluate)
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.2, random_state=42, stratify=y_data
)
print(f"Train Vector Matrix Size: {X_train.shape}")


print("\n[STAGE 03/05] Tuning and Training Optimized Random Forest Core Classifier Engine...")
# Hyperparameters configuration
intelligence_model = RandomForestClassifier(
    n_estimators=150,
    max_depth=25,
    min_samples_split=4,
    min_samples_leaf=2,
    n_jobs=-1,
    random_state=42
)

# multidimensional matrices
intelligence_model.fit(X_train, y_train)

# Statistical Scoring Matrix Generations
y_pred = intelligence_model.predict(X_test)
final_accuracy = accuracy_score(y_test, y_pred) * 100

print("\n======================= VALIDATED INDUSTRIAL PERFORMANCE SUMMARY =======================")
print(f"System Global Accuracy Benchmark Score: {final_accuracy:.2f}%")
print("----------------------------------------------------------------------------------------")
print(classification_report(y_test, y_pred, target_names=classes_list))


print("[STAGE 04/05] Rendering Structural Confusion Matrix Data Validation Table...")


print(" TEXT CONFUSION MATRIX GRAPH LAYOUT")

cm = confusion_matrix(y_test, y_pred)
print(f"{'Actual / Pred':<15} | " + " | ".join([f"{c:<10}" for c in classes_list]))
print("-" * 55)
for idx, row in enumerate(cm):
    print(f"{classes_list[idx]:<15} | " + " | ".join([f"{val:<10}" for val in row]))
print("=====================================================================\n")


print("[STAGE 05/05] Launching Production Level Enterprise UI Dashboard Server Application...")
# Initialize secure verification geocoding connection engine
geolocator = Nominatim(user_agent="industrial_taxonomy_intelligence_analytics_suite_v4")

month_analytics_map = {
    "January ( Winter)": 1, "February ( Winter)": 2, "March (spring)": 3,
    "April (Spring)": 4, "May ( Summer)": 5, "June (Summer)": 6,
    "July ( Monsoon)": 7, "August ( Monsoon)": 8, "September (Autumn)": 9,
    "October (Autumn)": 10, "November (Winter)": 11, "December ( Winter)": 12
}

def execute_predictive_inference_service(location_query, start_period, end_period):
    try:
        start_idx = month_analytics_map[start_period]
        end_idx = month_analytics_map[end_period]

        if start_idx <= end_idx:
            temporal_range = list(range(start_idx, end_idx + 1))
        else:
            temporal_range = list(range(start_idx, 13)) + list(range(1, end_idx + 1))

        # Geocoding Node Processing: Raw user input text mapping to physical location float vectors
        geospatial_vector = geolocator.geocode(location_query)
        if geospatial_vector is None:
            return (
                f"### ❌ System Alert: Location Resolution Failed\n"
                f"The target identifier **'{location_query}'** could not be resolved. "
                f"Please verify geographical spelling or input a recognized administrative region/city name."
            )

        lat = geospatial_vector.latitude
        lon = geospatial_vector.longitude
        abs_lat = abs(lat)

        # Core Inference Loops scoring over the requested custom temporal range boundary window
        probability_matrix = []
        for month_token in temporal_range:
            feature_input = np.array([[float(lat), float(lon), int(month_token), float(abs_lat)]])
            prediction_probabilities = intelligence_model.predict_proba(feature_input)[0]
            probability_matrix.append(prediction_probabilities)

        # Calculate calculated mathematical mean array
        aggregated_probabilities = np.mean(probability_matrix, axis=0)
        optimal_class_index = np.argmax(aggregated_probabilities)
        validated_kingdom = label_encoder.inverse_transform([optimal_class_index])[0]

        # Formulate Formal Production Analytical Report Document Screen Output
        report_output = f"### 📍 Geospatial Metadata Resolution\n"
        report_output += f"- **Resolved Address:** {geospatial_vector.address}\n"
        report_output += f"- **Target Coordinates:** Latitude: `{lat:.5f}°`, Longitude: `{lon:.5f}°`\n"
        report_output += f"- **Temporal Evaluation Window:** {start_period} to {end_period} ({len(temporal_range)} Months Aggregated)\n\n"

        report_output += f"### 🤖 Predictive Analytics Result\n"
        report_output += f"#### Optimal Biological Classification: **{validated_kingdom.upper()}**\n"
        report_output += f"Based on spatial-temporal pattern matching, the ecosystem within this boundary shows a high-probability distribution correlation towards the **{validated_kingdom}** taxonomy group during the selected period.\n\n"

        report_output += "### 📈 Statistical Confidence Breakdown (Mean Probability)\n"
        for classification_index, confidence_score in enumerate(aggregated_probabilities):
            taxonomy_label = label_encoder.inverse_transform([classification_index])[0]
            percentage = confidence_score * 100
            report_output += f"- **{taxonomy_label:<10}** : `{percentage:.2f}%` Confidence\n"

        return report_output

    except Exception as operational_error:
        return f"### ⚠️ System Operational Exception\nAn unexpected error occurred during model evaluation. Details: `{str(operational_error)}`"

# --- Building High-End Premium Corporate UI Panel Theme Layout ---
month_collection = list(month_analytics_map.keys())

with gr.Blocks(theme=gr.themes.Soft(primary_hue="teal", secondary_hue="slate")) as production_platform_ui:

    gr.Markdown("# 🌍 TAXONOMY INTELLIGENCE & GEOSPATIAL ANALYTICS PLATFORM")
    gr.Markdown("#### Predictive Modeling Interface | Core Infrastructure Framework: Polars & Random Forest Classifier")
    gr.Markdown("---")

    # Active KPIs Metrics Monitoring Indicators Row Block
    with gr.Row():
        gr.Textbox(value=f"{final_accuracy:.2f}%", label="Validated Model Accuracy Score", interactive=False)
        gr.Textbox(value="Geospatial-Temporal Matrix Engine", label="Inference Features Matrix Type", interactive=False)
        gr.Textbox(value="Random Forest Ensemble Optimized Core", label="Pipeline Backend Architecture Architecture", interactive=False)

    gr.Markdown("### 🔍 Spatial-Temporal Analytical Parameters")

    with gr.Row():
        # Configuration Controls Inputs Sidebar
        with gr.Column(scale=4):
            gr.Markdown("##### 🛠️ Geographic & Timeline Specifications")

            input_location_string = gr.Textbox(
                label="Geographical Target Location Identifier",
                placeholder="Enter country, city, continent or ecological zone (e.g., London, Amazon, Tokyo)...",
                value="New Delhi"
            )

            with gr.Row():
                input_start_window = gr.Dropdown(choices=month_collection, label="Commencement Month", value="January (Q1 - Winter)")
                input_end_window = gr.Dropdown(choices=month_collection, label="Termination Month", value="April (Q2 - Spring)")

            execute_analytics_trigger = gr.Button("RUN INFRASTRUCTURE INFERENCE 🚀", variant="primary")

        # Realtime Report Viewport Panel Terminal Output Display
        with gr.Column(scale=5):
            gr.Markdown("##### 🖥️ Live Platform Analytics Engine Output")
            analytical_output_viewport = gr.Markdown(value="*System initialized. Awaiting parameters submission. Configure input specifications and execute inference to generate data matrix.*")

    execute_analytics_trigger.click(
        fn=execute_predictive_inference_service,
        inputs=[input_location_string, input_start_window, input_end_window],
        outputs=analytical_output_viewport
    )

print("[SUCCESS] All system parameters compiled. Booting local network web app client interfaces...")
# Activate web streaming endpoints tunnels
production_platform_ui.launch(share=True, debug=True)
```
